```python
from datetime import datetime, timedelta

class SurvivalDuration:
    def __init__(self, start_time: datetime, end_time: datetime = None):
        """
        Initialize the SurvivalDuration object with a start time.
        Optionally, an end time can be provided.

        Parameters:
        - start_time (datetime): The starting point for survival calculation.
        - end_time (datetime, optional): The ending point. If not provided, assumes survival is ongoing.
        """
        self.start_time = start_time
        self.end_time = end_time

    def calculate_duration(self) -> timedelta:
        """
        Calculates the duration between start_time and end_time.
        If end_time is not set, it uses the current time as end_time.

        Returns:
        - timedelta: The duration between start and end times.
        """
        if self.end_time is None:
            current_time = datetime.now()
            return current_time - self.start_time
        return self.end_time - self.start_time

    def set_end_time(self, end_time: datetime):
        """
        Sets the end time for survival duration.

        Parameters:
        - end_time (datetime): The ending point for survival calculation.
        """
        self.end_time = end_time

# Example usage
start = datetime(2023, 10, 1, 8, 0, 0)
survival = SurvivalDuration(start)
print(f"Survival time so far: {survival.calculate_duration()}")

# Setting an end time
end = datetime(2023, 10, 5, 14, 30, 0)
survival.set_end_time(end)
print(f"Total survival time: {survival.calculate_duration()}")
```

```python
from datetime import datetime, timedelta
import math

class SurvivalDuration:
    def __init__(self, start_time: datetime):
        """
        Initializes the SurvivalDuration class with a start time.

        Parameters:
        - start_time (datetime): The time when survival begins.
        """
        self.start_time = start_time
        self.end_time = None  # End time is set to None by default

    def set_end_time(self, end_time: datetime):
        """
        Sets the end time for the survival period.

        Parameters:
        - end_time (datetime): The time when survival ends.
        """
        self.end_time = end_time

    def calculate_duration(self) -> timedelta:
        """
        Calculates the total duration as a timedelta.
        If the end time is not set, it calculates the duration up to the current time.

        Returns:
        - timedelta: The duration between the start and end times.
        """
        if self.end_time is None:
            return datetime.now() - self.start_time
        return self.end_time - self.start_time

    def calculate_duration_in_units(self) -> dict:
        """
        Calculates the duration in six different units: seconds, minutes, hours, days, weeks, and months.

        Returns:
        - dict: A dictionary containing the duration in different units.
        """
        duration = self.calculate_duration()
        duration_in_seconds = duration.total_seconds()

        # Converting duration into various units
        seconds = duration_in_seconds
        minutes = duration_in_seconds / 60
        hours = duration_in_seconds / 3600
        days = duration_in_seconds / (3600 * 24)
        weeks = duration_in_seconds / (3600 * 24 * 7)
        months = duration_in_seconds / (3600 * 24 * 30.44)  # Using average month length (30.44 days)

        return {
            "seconds": seconds,
            "minutes": minutes,
            "hours": hours,
            "days": days,
            "weeks": weeks,
            "months": months
        }

# Example usage
start_time = datetime(2024, 10, 1, 12, 0, 0)  # Survival started at noon on Oct 1, 2024
end_time = datetime(2024, 10, 17, 12, 0, 0)  # Survival ended at noon on Oct 17, 2024

survival = SurvivalDuration(start_time)
survival.set_end_time(end_time)

# Get duration in different units
duration_units = survival.calculate_duration_in_units()

# Print the results
for unit, value in duration_units.items():
    print(f"Duration in {unit}: {value}")
```
